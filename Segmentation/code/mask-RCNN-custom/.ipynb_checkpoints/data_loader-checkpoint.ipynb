```python
class cocoSeg(Dataset):
    def __init__(self, data_dir=None, annots_dir=None, transform=None, test_transform=None):
        if not (data_dir==None or annots_dir==None):
            self.data_dir = data_dir
            self.annots_dir = annots_dir
            self.ds = coco_train = dset.CocoDetection(root=self.data_dir,
                                annFile=self.annots_dir)
        else:
            print("Please Specify Directory")
            
        self.n_classes = len(self.ds.coco.cats)
    
    def __len__(self):
        return self.ds.__len__()
    
    def __getitem__(self, idx):
        img, targ = self.ds.__getitem__(idx)
        labels, masks, boxes, image_ids, iscrowd, area = [], [], [], [], [], []
        for item in targ:
            label = item['category_id']
            segmentation = mask.frPyObjects(item['segmentation'], img.size[1], img.size[0])
            bin_mask = mask.decode(segmentation)[:,:,0]
            obj_pixels = np.where(bin_mask)
            print(obj_pixels)
            xmin = np.min(obj_pixels[1])
            xmax = np.max(obj_pixels[1])
            ymin = np.min(obj_pixels[0])
            ymax = np.max(obj_pixels[0])
            if (((xmax-xmin)<=10) | (ymax-ymin)<=10):
                xmax = xmin+10
                ymax = ymin+10
            boxes.append([xmin, ymin, xmax, ymax])

            iscrowd.append(item['iscrowd'])
            area.append(item['area'])
            labels.append(label)
            masks.append(bin_mask)
            image_ids.append(item["image_id"])
            
        target = {}
        target["boxes"] = torch.tensor(np.array(boxes, dtype=int), dtype=torch.float32)
        target["labels"] = torch.as_tensor(labels, dtype=torch.int64)
        target["masks"] = torch.tensor(np.array(masks), dtype=torch.float32)
        target["image_id"] = torch.tensor(image_ids, dtype=torch.int64)
        target["area"] = torch.tensor(area, dtype=torch.float32)
        target["iscrowd"] = torch.tensor(iscrowd, dtype=torch.torch.int64)
        img = torch.tensor(np.array(img.getdata()).reshape(img.size[1], img.size[0], 3), dtype=torch.float32).permute(2, 0, 1)
        return img, target
        
    def get_category(self, label):
        return self.ds.coco.cats[label]["name"]
    
    def collate_fn(self, batch):
        return tuple(zip(*batch))
```

# Build the Model

```python
def get_model_instance_segmentation(num_classes):
    # load an instance segmentation model pre-trained pre-trained on COCO
    model = torchvision.models.detection.maskrcnn_resnet50_fpn(pretrained=True)

    # get number of input features for the classifier
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    # replace the pre-trained head with a new one
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

    # now get the number of input features for the mask classifier
    in_features_mask = model.roi_heads.mask_predictor.conv5_mask.in_channels
    hidden_layer = 256
    # and replace the mask predictor with a new one
    model.roi_heads.mask_predictor = MaskRCNNPredictor(in_features_mask,
                                                       hidden_layer,num_classes)
    return model
```

```python
model = get_model_instance_segmentation(train_dataset.n_classes).to(device)
```

```python
params = [p for p in model.parameters() if p.requires_grad]
optimizer = torch.optim.SGD(params, lr=0.005,
                            momentum=0.9, weight_decay=0.0005)
# and a learning rate scheduler
lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer,
                                                step_size=3,
                                                gamma=0.1)
```

```python
def rle2bbox(rle, shape):
    '''
    rle: run-length encoded image mask, as string
    shape: (height, width) of image on which RLE was produced
    Returns (x0, y0, x1, y1) tuple describing the bounding box of the rle mask
    
    Note on image vs np.array dimensions:
    
        np.array implies the `[y, x]` indexing order in terms of image dimensions,
        so the variable on `shape[0]` is `y`, and the variable on the `shape[1]` is `x`,
        hence the result would be correct (x0,y0,x1,y1) in terms of image dimensions
        for RLE-encoded indices of np.array (which are produced by widely used kernels
        and are used in most kaggle competitions datasets)
    '''
    
    a = np.fromiter(rle.split(), dtype=np.uint)
    a = a.reshape((-1, 2))  # an array of (start, length) pairs
    a[:,0] -= 1  # `start` is 1-indexed
    
    y0 = a[:,0] % shape[0]
    y1 = y0 + a[:,1]
    if np.any(y1 > shape[0]):
        # got `y` overrun, meaning that there are a pixels in mask on 0 and shape[0] position
        y0 = 0
        y1 = shape[0]
    else:
        y0 = np.min(y0)
        y1 = np.max(y1)
    
    x0 = a[:,0] // shape[0]
    x1 = (a[:,0] + a[:,1]) // shape[0]
    x0 = np.min(x0)
    x1 = np.max(x1)
    
    if x1 > shape[1]:
        # just went out of the image dimensions
        raise ValueError("invalid RLE or image dimensions: x1=%d > shape[1]=%d" % (
            x1, shape[1]
        ))

    return x0, y0, x1, y1
```
